```python
import os
from matplotlib import pyplot as plt
```

```python
class BasicDataVisualizer:
    x = []
    y = []
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def plot(self):
        plt.plot(self.x, self.y)
        plt.show()
```

```python
# Dummy Data
x = [1,2,3,4,5]
y = [10,2,3,4,5]
```

```python
dataVisualizer = BasicDataVisualizer(x,y)
```

```python
dataVisualizer.plot()
```

```python
import json

class JsonDataVisualizer(BasicDataVisualizer):
    x = []
    y = []
    
    def __init__(self, jsonFileName):
        f = open(jsonFileName)
        jsonData = json.load(f)
        x = []
        y = []
        for datum in jsonData['data']:
            x.append(datum["x"])
            y.append(datum["y"])
        
        super().__init__(x, y)
        f.close()
```

```python
jsonDataVisualizer = JsonDataVisualizer("sample.json")
```

```python
jsonDataVisualizer.plot()
```
